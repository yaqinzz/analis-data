### Cleaning Data

#### Menghapus colum yang tidak diperlukan

```python
hour_df.drop(columns=['instant', 'windspeed'], inplace=True)
day_df.drop(columns=['instant', 'windspeed'], inplace=True)
```

**Insight:**
- menghapus colum instant dan windspeed karena tidak ada pertanyaan bisnis terkait kecepatan angin terhadap jumlah penyewa sepeda.


#### Mengubah Tipe Data

```python
columns = ['season', 'mnth', 'holiday', 'weekday', 'weathersit']
 
for column in columns:
    hour_df[column] =  hour_df[column].astype("category")
    day_df[column] =  day_df[column].astype("category")
```

```python
hour_df['dteday'] = pd.to_datetime(hour_df['dteday'])
day_df['dteday'] = pd.to_datetime(day_df['dteday'])
```

```python
hour_df.rename(columns={
    'yr':'year','mnth':'month','weekday':'one_of_week', 'weathersit':'weather_situation', 'windspeed':'wind_speed','cnt':'count_cr','hum':'humidity'
    },inplace=True)
```

```python
day_df.rename(columns={
    'yr':'year','mnth':'month','weekday':'one_of_week', 'weathersit':'weather_situation', 'windspeed':'wind_speed','cnt':'count_cr','hum':'humidity'
    },inplace=True)
```

```python
hour_df['month'] = hour_df['month'].map({
    1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun',
    7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'
})
hour_df['season'] = hour_df['season'].map({
    1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'
})
hour_df['one_of_week'] = hour_df['one_of_week'].map({
    0: 'Sunday', 1: 'Monday', 2: 'Tuesday', 3: 'Wednesday', 4: 'Thursday', 5: 'Friday', 6: 'Saturday'
})
hour_df['weather_situation'] = hour_df['weather_situation'].map({
    1: 'Clear',
    2: 'Misty',
    3: 'Light Snow',
    4: 'Severe Weather'
})
hour_df["year"] = hour_df["year"].map({0: 2011, 1: 2012})
```

```python
day_df['month'] = day_df['month'].map({
    1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun',
    7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'
})
day_df['season'] = day_df['season'].map({
    1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'
})
day_df['one_of_week'] = day_df['one_of_week'].map({
    0: 'Sunday', 1: 'Monday', 2: 'Tuesday', 3: 'Wednesday', 4: 'Thursday', 5: 'Friday', 6: 'Saturday'
})
day_df['weather_situation'] = day_df['weather_situation'].map({
    1: 'Clear',
    2: 'Misty',
    3: 'Light Snow',
    4: 'Severe Weather'
})
day_df["year"] = day_df["year"].map({0: 2011, 1: 2012})
```

**Insight:**
- Kolom 'instant' dan 'windspeed' dihapus karena tidak ada pertanyaan bisnis terkait kecepatan angin terhadap jumlah penyewa sepeda.
- Tipe data pada kolom 'season', 'mnth', 'holiday', 'weekday', dan 'weathersit' diubah menjadi tipe data kategori untuk memudahkan analisis.
- Kolom 'dteday' diubah menjadi tipe data datetime untuk memudahkan manipulasi tanggal.
- Nama kolom diubah untuk meningkatkan keterbacaan dan konsistensi.
- Nilai pada kolom 'month', 'season', 'one_of_week', 'weather_situation', dan 'year' diubah menjadi bentuk yang lebih mudah dipahami.

## Visualization & Explanatory Analysis

### Pertanyaan 1: Seberapa besar pengaruh cuaca terhadap jumlah peminjaman sepeda?

```python
colors = ["#72BCD4", "#D3D3D3", "#D3D3D3", "#D3D3D3"]
plt.figure(figsize=(10,6))
sns.barplot(
    x='weather_situation',
    y='count_cr',
    hue='weather_situation',
    data=hour_df,
    palette=colors,
    legend=False
    )

plt.title('Jumlah Pengguna Sepeda berdasarkan Kondisi Cuaca')
plt.xlabel('Kondisi Cuaca')
plt.ylabel('Jumlah Pengguna Sepeda')
plt.show()
```

```python
colors = ["#72BCD4", "#D3D3D3", "#D3D3D3"]
plt.figure(figsize=(10,6))
sns.barplot(
    x='weather_situation',
    y='count_cr',
    hue='weather_situation',
    data=day_df,
    palette=colors,
    legend=False
    )

plt.title('Jumlah Pengguna Sepeda berdasarkan Kondisi Cuaca')
plt.xlabel('Kondisi Cuaca')
plt.ylabel('Jumlah Pengguna Sepeda')
plt.show()
```

**Insight:**
- Cuaca yang cerah memiliki jumlah penyewaan sepeda tertinggi dibandingkan dengan kondisi cuaca lainnya.
- Kondisi cuaca yang buruk seperti hujan salju ringan dan cuaca buruk parah memiliki jumlah penyewaan sepeda yang jauh lebih rendah.

### Pertanyaan 2: Apakah penggunaan sepeda lebih tinggi pada hari kerja atau akhir pekan?

```python
colors = ["#72BCD4", "#D3D3D3"]
plt.figure(figsize=(10,6))
sns.barplot(
    x='workingday',
    y='count_cr',
    hue='workingday',
    data=hour_df,
    palette=colors,
    legend=False
    )

plt.title('Jumlah Pengguna Sepeda berdasarkan Hari Kerja')
plt.xlabel('Hari Kerja')
plt.ylabel('Jumlah Pengguna Sepeda')
plt.show()
```

```python
colors = ["#72BCD4", "#D3D3D3"]
plt.figure(figsize=(10,6))
sns.barplot(
    x='workingday',
    y='count_cr',
    hue='workingday',
    data=day_df,
    palette=colors,
    legend=False
    )

plt.title('Jumlah Pengguna Sepeda berdasarkan Hari Kerja')
plt.xlabel('Hari Kerja')
plt.ylabel('Jumlah Pengguna Sepeda')
plt.show()
```

**Insight:**
- Penggunaan sepeda lebih tinggi pada hari kerja dibandingkan dengan akhir pekan.
- Hal ini menunjukkan bahwa banyak orang menggunakan sepeda sebagai alat transportasi untuk bekerja atau beraktivitas pada hari kerja.


### Pertanyaan ke 3: Apakah ada perbedaan signifikan dalam jumlah penyewaan sepeda antara musim semi, panas, gugur dan dingin?

```python
colors = ["#D3D3D3", "#D3D3D3", "#72BCD4", "#D3D3D3"]
plt.figure(figsize=(10,6))
sns.barplot(
    x='season',
    y='count_cr',
    hue='season',
    data=hour_df,
    palette=colors,
    legend=False
    )

plt.title('Jumlah Pengguna Sepeda berdasarkan Musim')
plt.xlabel('Musim')
plt.ylabel('Jumlah Pengguna Sepeda')
plt.show()
```

```python
colors = ["#D3D3D3", "#D3D3D3", "#72BCD4", "#D3D3D3"]
plt.figure(figsize=(10,6))
sns.barplot(
    x='season',
    y='count_cr',
    hue='season',
    data=day_df,
    palette=colors,
    legend=False
    )

plt.title('Jumlah Pengguna Sepeda berdasarkan Musim')
plt.xlabel('Musim')
plt.ylabel('Jumlah Pengguna Sepeda')
plt.show()
```

**Insight:**
- Terdapat perbedaan signifikan dalam jumlah penyewaan sepeda antara musim semi, panas, gugur, dan dingin.
- Musim gugur memiliki jumlah penyewaan sepeda tertinggi dibandingkan dengan musim lainnya.
- Musim semi memiliki jumlah penyewaan sepeda terendah dibandingkan dengan musim lainnya.
